```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
from nltk.stem.porter import *
import numpy as np
np.random.seed(2018)
from numpy import dot
from numpy.linalg import norm
```

```python

import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
import numpy as np
import cv2
import gc
```

```python
test_df = pd.read_csv('../input/shopee-product-matching/test.csv')
test_df
```

```python
test_image_list = test_df['image'].to_list()
test_image_list
```

# Image loading & resizing

```python
test_res = []
base_path = '../input/shopee-product-matching/test_images/'
for image in test_image_list:
#     print(base_path+image)
    test_image = base_path+image
    img = cv2.imread(test_image)
    test_res.append(cv2.resize(img, (32,32)))
    del img
_ = gc.collect()
```

```python
test_res
```

```python
x_test_arr = np.asarray(test_res)
x_test_arr.shape
```

# Data normalization

```python
x_test_arr = x_test_arr/255
```

# Load model

```python
# os.listdir("../input/cnn-sivyati-model-25")
# new_model = '/kaggle/input/models/cnn_sivyati_model.h5'
```

```python
os.listdir("../input")
```

```python
from tensorflow.keras.models import load_model
model = load_model('../input/cnn-sivyati-model-32/cnn_sivyati_model_32.h5')
```

# Run model

```python
y_pred_new = model.predict(x_test_arr)
submission_df = test_df
submission_df['pred'] = [np.argmax(element) for element in y_pred_new]
```

```python
drop_cols = ['image','image_phash','title']
submission_df = submission_df.drop(columns=drop_cols)
```

```python
submission_df['matches'] = submission_df.groupby(['pred'])['posting_id'].transform(lambda x : ' '.join(x))
```

```python
submission_df = submission_df.drop(columns='pred')
submission_df
```

```python
submission_df.to_csv('submission.csv', index = False)
```
